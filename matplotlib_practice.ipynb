```python
# import
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create the plot
plt.plot(x_axis, y_axis)

plt.show()
```

```python
# create plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)

plt.show()
```

```python
# create plot with ax.plt()
fig = plt.figure() # To change figure-level attributes (such as axis labels, a title, or a legend) or save the figure as an image
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)

plt.show()
```

```python
# create a quick graph using the object-oriented interface method
ax = plt.axes()
ax.plot(x_axis, y_axis)

plt.show()
```

```python
# create plot and add label for legend
plt.plot(x_axis, y_axis,
         marker = "*",
         color = '#CDB4DB',
         label = 'Boston')

# create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# set the y limit between 0 and 45
plt.ylim(0, 45)

# create title
plt.title("PyBer Fare by Month")

# add legend
plt.legend()

plt.show()
```

```python
# create plot and add label for legend
plt.plot(x_axis, y_axis,
         marker = "*",
         color = '#FFC8DD',
         linewidth = 2, 
         label = 'Boston')

# create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# set the y limit between 0 and 45
plt.ylim(0, 45)

# create title
plt.title("PyBer Fare by Month")

# add a grid
plt.grid()

# add legend
plt.legend()

# change figure size
plt.figure(figsize=(20, 10))

plt.show()
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, 
        y_axis, 
        color = '#FFAFCC',
        marker = 'd',
        linewidth = 2, 
        label = 'Boston')
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel('Month')
ax.set_ylabel('Fare ($)')
ax.set_ylim(0, 50)
ax.grid()

plt.show()
```

```python
# pie chart
plt.pie(y_axis, labels = x_axis)

plt.show()
```

```python
# assigning 12 colors 
colors = ['#FA6163', '#F46D52', '#F9864D', '#F58D3D', '#F9A339', '#F9C74F', '#90BE6D', '#5FBFA2', '#3FABF3', '#7E91F1', '#AB7CB6', '#F081AE']

#explode wedge
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# increase size
plt.subplots(figsize = (8, 8))  

plt.pie(y_axis, 
        explode = explode_values,
        colors = colors,
        labels = x_axis, 
        autopct = '%.1f%%') # "autopct" adding percentage, ".1f%" to 1 decimal place, "%.1f%%" formats as percentage

plt.show()
```

```python
# object oriented interface pie chart
fig, ax = plt.subplots()
ax.pie(y_axis, labels = x_axis)

plt.show()
```

```python
# object oriented interface pie chart
colors = ['#FA6163', '#F46D52', '#F9864D', '#F58D3D', '#F9A339', '#F9C74F', '#90BE6D', '#5FBFA2', '#3FABF3', '#7E91F1', '#AB7CB6', '#F081AE']

explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

fig, ax = plt.subplots()

ax.pie(y_axis, 
       labels = x_axis,
       colors = colors,
       autopct = '%.2f%%',
       radius = 2,
       explode = explode_values,
       startangle = 85,
       pctdistance = .8,
       labeldistance = 1.15)

ax.invert_xaxis()

plt.show()
```

```python
# error bar exercise
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
# cont, line graph with error bars
plt.errorbar(x_axis, 
             y_axis, 
             yerr = stdev, 
             capsize = 3)

plt.show()
```

```python
# same but object oriented approach
fig, ax = plt.subplots()
ax.errorbar(x_axis, 
            y_axis, 
            yerr = stdev, 
            capsize = 3)

plt.show()
```

```python
# error bars on bar chart
plt.bar(x_axis, 
        y_axis, 
        yerr = stdev, 
        capsize = 3)

plt.show()
```

```python
# bar error but object oriented
fig, ax = plt.subplots()
ax.bar(x_axis,
       y_axis,
       yerr = stdev,
       capsize = 3)

plt.show()
```

```python
# changing distance of major ticks on x-axis
import numpy as np

plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step = 5.0))
plt.gca().invert_yaxis()

plt.show()
```

```python
# same but object oriented
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step = 5.0))

plt.show()
```

```python
# adding minor ticks 
from matplotlib.ticker import MultipleLocator

# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize = (8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step = 5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))

plt.show()
```
